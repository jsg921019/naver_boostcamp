## Ⅰ. Line

* `axvline()` : 수직선

* `axhline()` : 수평선

```python
import numpy as np
import pandas as pd 
import matplotlib as mpl
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.set_aspect(1)
ax.axvline(0, color='red')
ax.axhline(0, color='green')
ax.axvline(-0.5, ymin=0.4, ymax=0.8, color='blue')
ax.axhline(-0.5, xmin=0.6, xmax=0.9, color='black')

ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)

plt.show()
```

```python
student = pd.read_csv('data/StudentsPerformance.csv')

fig, ax = plt.subplots(figsize=(10, 10), subplot_kw={'aspect':1})

math_mean = student['math score'].mean()
reading_mean = student['reading score'].mean()
color = ['royalblue' if m > math_mean and r > reading_mean else 'gray'
         for m, r in zip(student['math score'], student['reading score'])]

ax.axvline(math_mean, color='gray', linestyle='--')
ax.axhline(reading_mean, color='gray', linestyle='--')
ax.scatter(x=student['math score'], y=student['reading score'],
           color=color, zorder=10, alpha=0.5)

ax.set_xlabel('Math')
ax.set_ylabel('Reading')
ax.set_xlim(-3, 103)
ax.set_ylim(-3, 103)

plt.show()
```

## Ⅱ. Span

- axvspan
- axhspan

```python
fig, ax = plt.subplots()

ax.set_aspect(1)
ax.axvspan(0, 0.5, color='red')
ax.axhspan(0, 0.5, color='green')
ax.axvspan(0.25, 0.75, ymin=0.15, ymax=0.55, color='blue')
ax.axhspan(-0.25, 0.25, xmin=0.2, xmax=0.6, color='black')

ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw={'aspect':1})

ax.axvspan(-3, math_mean, color='gray', linestyle='--', zorder=0, alpha=0.3)
ax.axhspan(-3, reading_mean, color='gray', linestyle='--', zorder=0, alpha=0.3)
ax.scatter(x=student['math score'], y=student['reading score'],
           alpha=0.4, s=20, color=color, zorder=10)

ax.set_xlabel('Math')
ax.set_ylabel('Reading')
ax.set_xlim(-3, 103)
ax.set_ylim(-3, 103)

plt.show()
```
