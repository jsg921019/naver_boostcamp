## Ⅰ. Matplotlib

```python
import numpy as np 
import matplotlib.pyplot as plt
```

파이썬에서 쓰는 데이터 시각화 라이브러리에는 Matplotlib, Seaborn, Plotly, Bokeh, Altair가 있다.  
Matplotlib는 이중에서 범용성이 제일 넓고, base가 되는 라이브러리이다.  
Matplotlib은 두 가지 API를 따로 지원한다.

- Pyplot API : 순차적 방법
- 객체지향(Object-Oriented) API : 그래프에서 각 객체에 대해 직접적으로 수정하는 방법

이 노트북에서는 가능한 객체지향 API를 사용한다.

## Ⅱ. Figure와 Axes

Figures는 Axes를 담는 컨테이너이며 Axes는 그래프가 담기는 영역들이다.  
* Figure 는 `plt.figure()`로 생성하고 `figsize`로 크기를 설정할 수 있다.  
* Ax는 `fig.add_subplot()`으로 figure에 추가하고 (행 갯수, 열 갯수, 인덱스번호)를 지정하여 크기를 지정할 수 있다.

```python
fig = plt.figure(figsize=(12, 7))
ax1 = fig.add_subplot(2, 2, 1) 
ax2 = fig.add_subplot(2, 2, 3)
ax2 = fig.add_subplot(1, 2, 2)
plt.show()
```

## Ⅲ. 그래프 그리기

* Ax에 matplotlib의 여러 함수를 사용해서 그래프를 그릴 수 있다.

```python
fig = plt.figure()
ax = fig.add_subplot() 

ax.plot([1, 1, 1])
ax.plot([1, 2, 3])
ax.plot([3, 3, 3])

plt.show()
```

* `color` parameter로 그래프의 색을 직접 지정할 수 있다.

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot([1, 1, 1], color='r') # 한 글자로 정하는 색상
ax.plot([1, 2, 3], color='forestgreen') # color name
ax.plot([3, 3, 3], color='#000000') # hex code (BLACK)

plt.show()
```

* `label` parameter과 `ax.legend()`로 범례를 표시할 수 있다.

```python
fig = plt.figure()
ax = fig.add_subplot() 

ax.plot([1, 1, 1], label='1') 
ax.plot([2, 2, 2], label='2') 
ax.plot([3, 3, 3], label='3')

ax.legend()
plt.show()
```

* `ax.set_title` 으로 그래프에 제목을 표시할 수 있다.

```python
fig = plt.figure()
ax = fig.add_subplot() 

ax.plot([1, 1, 1]) 
ax.plot([2, 2, 2]) 
ax.plot([3, 3, 3])

ax.set_title('Basic Plot')
plt.show()
```

* `ax.set_xticks` 과 `ax.set_xticklabels` 로 축의 눈금정보를 변경할 수 있다.

```python
fig = plt.figure()
ax = fig.add_subplot() 

ax.plot([1, 1, 1]) 
ax.plot([2, 2, 2]) 
ax.plot([3, 3, 3])

ax.set_xticks([0, 1, 2])
ax.set_xticklabels(['zero', 'one', 'two'])

plt.show()
```

* `ax.text` 와 `ax.annotate` 으로 그래프에 text를 쓸 수 있다.

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot([1, 1, 1]) 
ax.plot([2, 2, 2]) 
ax.plot([3, 3, 3])

ax.text(x=0.3, y=2.05, s='This is Text', fontsize=15)
ax.annotate(text='This is Annotate', fontsize=12,
            xy=(0.9, 2.15), xytext=(1.2, 2.4),
            arrowprops=dict(facecolor='black', shrink=0.1))

plt.show()
```

## Ⅳ. Spines

- set_visible
- set_linewidth
- set_position
  - 'center' -> ('axes', 0.5)
  - 'zero' -> ('data', 0.0)

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for ax in axes:
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

axes[0].spines['left'].set_linewidth(1.5)
axes[0].spines['bottom'].set_linewidth(1.5)

axes[1].spines['left'].set_position('center')
axes[1].spines['bottom'].set_position('center')

axes[2].spines['left'].set_position(('data', 0.3))
axes[2].spines['bottom'].set_position(('axes', 0.2))

plt.show()
```
